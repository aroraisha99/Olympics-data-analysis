```python
games_athletes = athlete.pivot_table(athlete, index=['Games'], aggfunc=lambda x: len(x.unique())).reset_index()[['Games','ID']]
fig, ax = plt.subplots(figsize=(22,6))
a = sns.barplot(x='Games', y='ID', data=games_athletes, ax=ax, color="#2196F3")
a.set_xticklabels(labels=games_athletes['Games'],rotation=90)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')

ax.set_xlabel('Olympic Game', size=14, color="#0D47A1")
ax.set_ylabel('Number of Athletes', size=14, color="#0D47A1")
ax.set_title('Athletes in each Olympic game', size=18, color="#0D47A1")

plt.show()

#Number of athletes in Summer Olympics was maximum in the year 2008.
```

```python
gold_medal = athlete[athlete['Medal']=='Gold']
```

```python
from scipy.stats import ttest_ind
#does sports depend on age
#NULL HYPOTHESIS:Sports does not depend on age
#ALTERNATE HYPOTEHSIS:Sports does depend on age
Age_mean = df2['Age'].mean()
Sport_mean = df2['Sport'].mean()
print("Age mean value:",Age_mean)
print("Sport mean value:",Sport_mean )
Age_std =df2['Age'].std()
Sport_std =df2['Sport'].std()
print("Age std value:",Age_std)
print("Sport std value:",Sport_std)
ttest,pval = ttest_ind(df2['Sport'],df2['Age'])
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")


#Sports does depend on Age
```

```python
#does team depend on sport
#NULL HYPOTHESIS:Sports does not depend on team
#ALTERNATE HYPOTEHSIS:Sports does depend on team

Team_mean = df2['Team'].mean()
Sport_mean = df2['Sport'].mean()
print("Team mean value:",Team_mean)
print("Sport mean value:",Sport_mean)
Team_std =df2['Team'].std()
Sport_std =df2['Sport'].std()
print("Team std value:",Team_std)
print("Sport std value:",Sport_std)
ttest,pval = ttest_ind(df2['Team'],df2['Sport'])
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")


#Sports does not depend on Team
```
